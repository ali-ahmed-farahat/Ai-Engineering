```python
# I will create a dictionary with the numbers that are written as words
dict_str_to_int = {'forty':40, 'thirty':30}
df["time_spent_minutes"] = df["time_spent_minutes"].apply(lambda x: dict_str_to_int[x] if isinstance(x, str) and x in dict_str_to_int else x)
df
```

```python
df["time_spent_minutes"]  = df["time_spent_minutes"].astype(float)

mean = df["time_spent_minutes"].mean()
df["time_spent_minutes"] = df["time_spent_minutes"].fillna(value = mean)
df["time_spent_minutes"]
```

```python
#let's convert it to 2 decimal places
df["time_spent_minutes"] = df["time_spent_minutes"].round(2)
```

```python
#now let's drop the row where date of interaction is not availible since it's just one
df.dropna(subset = ["date_of_interaction"], inplace = True)
df.head()
```

```python
df["date_of_interaction"] = pd.to_datetime(df["date_of_interaction"])
```
